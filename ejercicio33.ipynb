## Explicación de código en notebook 3.33

```python
a = 3
b = a               # b es 3
print("b es ", b)

a = 3
b = a
a = 'spam'          # b es 3
print("b es ", b)
```

En este código la variable "b" no se define como "a", si no que se define como el valor de "a" a la hora de definir "b = a". Esto sucede porque lo que hace es crear unnuevo espacio en memoria con el nuevo valor de "a" y "b" sigue refernenciando al anterior valor.



```python
L = [2, 3, 4]	    # A mutable object
M = L			    # Make a reference to the same object
L[0] = 24 		    # An in-place change

print("L = ", L)    # [24, 3, 4]
print("M = ", M)    # [24, 3, 4]
```

En este caso se cambia el primer valor de la lista de la varible "L". "M" también cambia porque al hacer "L=[0]" cambia el contenido en vez de remplazar la variable con otra.

```python
L = [2, 3, 4]
M = L[:]		    # Make a copy of L (or list(L), copy.copy(L), etc.)
L[0] = 24

print("L = ", L)    # [24, 3, 4]

print("M = ", M)    # [2, 3, 4]  M is not changed
```

Ahora cambia el resultado porque se hace una copia del contenido en la varibale "L" en vez de señalar la posición de memoria en la que se encuentra. 

```python
L = [1, 2, 3]
M = L 			    # M and L reference the same object

L == M 			    # True: Same values => operador igualdad de valores
L is M 			    # True: operador identidad de objetos: compara las referencias (los punteros)
```

Este ejemplo trata de los operadores de "==" y "is". En el caso de "==" comprueba si los contenidos son iguales, y "is" comprueba si están en la misma posición de memoria.

```python


L = [1, 2, 3]
M = [1, 2, 3]	    # M and L reference different objects

print(
    "L == M",
    L == M 			# True: Same values
)

print(
    "L is M", 	    # False: operador identidad de objetos: compara las referencias (los punteros)
    L is M
)
```

En este ejemplo se ve mucho mejor

```python
X = 42
Y = 42		    # Should be two different objects

X  == Y         # True
X is Y 		    # True: Same object anyhow: caching at work!!
```

Python en algunos casos, cuando son pequeños strings o integrales los reutiliza para usar los menos espacios de variables posibles. Por eso aparece que es True "X is Y"
