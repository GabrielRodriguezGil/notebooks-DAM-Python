## Ejercicio de explicación del 3.34

```python

def f(a):	    	# a is assigned to (references) the passed object
	a = 99 			# Changes local variable a only

b = 88		
f(b)				# a and b both reference same 88 initially
print("b = ", b)	# b is not changed
```

En este programa se cambia el valor de "a", pero solo en la función. Por lo que a la hora de pedir "a" fuera de la función sgue teniendo su valor original.

```python


def changer(a, b):	    # Arguments assigned references to objects
	a = 2			# Changes local name's value only
	b[0] = 'spam' 	# Changes shared object in place	
					
x = 1
l = [1, 2]
changer(x, l)           # Caller: Pass immutable and mutable objects
print("l = ", l)        # ['spam', 2]  # x is unchanged, l is different!
assert x == 1, "x es 1"
```

En el caso de las listas es diferente, si que cambia el contenido original a la hora de seleccionar valores de la lista.

```python
l = [1, 2]
changer(x, l[:]) 	# Pass a copy, so our 'l' does not change 

print("l = ", l) 
```

En este caso se crea una copia de la lista, por lo que no cambia su valor.

```python
l = [1, 2]
def changer(a, b):
    b = b[:]		# Copy input list so we don't impact caller
    a = 2
    b[0] = 'spam'	# Changes our list copy only

changer(x, l)
assert l == [1, 2], "l es una invariante"
print("l = ", l) 
```

En este caso la copia la hace dentro de la función

```python


l = [1, 2]
def changer(a, b):	   
		a = 2			
		b[0] = 'spam'   # TypeError: 'tuple' object does not support item assignment

changer(x, tuple(l))	# Pass a tuple, so changes are errors

print("l = ", l)
assert l == [1, 2], "l es una invariante"

```

En este ejemplo lanza error porque lo convierte en tupla. Las tuplas son invariables, es decir, que no se puede modificar

```python
def f(a, b, c): 
    print(a, b, c)

f(1, 2, 3) 
```

Las funciones asumen los valores en el orden en el que lo recibe

```python

def f(a, b, c): 
    print(a, b, c)

f(c=3, b=2, a=1)    # 1 2 3
```

```python
f(1, c=3, b=2)		# a gets 1 by position, b and c passed by name
                    # 1 2 3

# they make your calls a bit more self documenting
# f(name='Bob', age=40, job='dev')
```

En cambio, si a las varibles de la función se le asumen valores, la función usará los valores que tengan asignados

```python
def f(a, b=2, c=3): 	# a required, b and c optional
	print(a, b, c)


f(1)        # 1 2 3
f(a=1)      # 1 2 3
```

```python
def f(a, b=2, c=3): 	# a required, b and c optional
	print(a, b, c)
    
f(1, 4)     # 1 4 3
f(1, 4, 5)	# Override defaults: 1 4 5

f(1, c=6)	
```

En estos casos la función usa valores predeterminados en "b" y "c". Si no asumes valores para ellos usará las de por defecto sino las cambiará.

```python
def func(spam, eggs, toast=0, ham=0):
	print((spam, eggs, toast, ham)) 	# First 2 required

func(1, 2)							# Output: (1, 2, 0, 0)
func(1, ham=1, eggs=0)				# Output: (1, 0, 0, 1)
func(spam=1, eggs=0)				# Output: (1, 0, 0, 0)
func(toast=1, eggs=2, spam=3)		# Output: (3, 2, 1, 0)
func(1, 2, 3, 4) 		
```

El tercer y cuarto valor de la función no son obligatorios, pero como cualquier otro se puede modificar su contenido si se le asume otro al llamar a la función

```python
def kwonly(a, *b, c):
    print(a, b, c)

kwonly(1, 2, c=3)       # 1 (2,) 3

kwonly(a=1, c=3)        # 1 () 3

kwonly(1, 2, 3)         # TypeError: kwonly() missing 1 required keyword-only argument: 'c'
```

En python se puede usar "*" para indicar que esa varible no es obligatoria, y recibirá los valores en forma de lista. En el último caso lanza error porque no se le asume valor a "c", porque b asume (2,3).

```python


def kwonly(a, *, b='spam', c='ham'):
	print(a, b, c)

kwonly(1)               # 1 spam ham
kwonly(1, c=3)          # 1 spam 3
kwonly(a=1)             # 1 spam ham
kwonly(c=3, b=2, a=1)   # 1 2 3
kwonly(1, 2)            # TypeError: kwonly() takes 1 positional argument but 2 were given

```

También se puede usar "*" de forma que asuma valores para que sea obligatorio definir las varibles de la función.

```python
def f(*args):
    print(args)

f()                 # ()
f(1,)               # (1,)
f(1, 2, 3, 4)       # (1, 2, 3, 4)
```

Al usar "*" antes del nombre de la variable lo que hace es que asuma valores en forma de tupla(lista que no se puede modficar)

```python
def f(**args):
    print(args)
f()                 # {}
f(a=1, b=2)         # {'a': 1, 'b': 2}
```

Si se usa "**" en vez de er tuplas son diccionarios

```python
def func(a, b, c, d): print(a, b, c, d)

args = (1, 2)
args += (3, 4)
func(*args) 
```

"*" pasa de varibles a tuplas y viceversa

```python
def func(a, b, c, d): print(a, b, c, d)


args = {'a': 1, 'b': 2, 'c': 3}
args['d'] = 4
func(**args)    # Same as func(a=1, b=2, c=3, d=4)

```

"**" pasa de varibles a diccionarios y viceversa
