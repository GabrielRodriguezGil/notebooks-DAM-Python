# Unified try Example 
---

Ejercecio del libro de Learning Python, que nos propone una serie de codigo, y explicar porque tiene esa salida. Utilizando `Try` , `Except` , `else` y `Finally`.

```python
sep = "-" * 45 + "\n"

print(sep + 'EXCEPTION RAISED AND CAUGHT')
try:
    x = 'spam'[99]
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

En este caso el except se ejecuta porque en el try se ejecuta `x = spam[99]` y spam no tiene longitud 99, con lo cual salta `IndexError` y entra en el except. Por ultimo el finally siempre se ejecuta, y el ultimo print es fuera del try, con lo cual se ejecutara tambien si o si 

```python
print(sep + 'NO EXCEPTION RAISED')
try:
    x = 'spam'[3]
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

Al contrario que en el caso anterior, al intentar acceder a una posicion del string que si existe, la excepcion no ocurre con lo cual el except no se ejecuta. Como antes el finally y el "after run" siempre se ejecutan

```python
print(sep + 'NO EXCEPTION RAISED, WITH ELSE')
try:
    x = 'spam'[3]
except IndexError:
    print('except run')
else:
    print("else run")
finally:
    print('finally run')
print('after run')
```

Como en el anterior ejemplo en el try no ocurre ninguna excepcion, con lo cual el except no se ejecuta, y el else (que solo se ejecuta si ninguno de los excepts se han ejecutado) se ejecuta. Y como siempre el finally y lo que no pertenece al try es ejecutado.

```python
print(sep + 'EXCEPTION RAISED BUT NOT CAUGHT')
try:
    x = 1 / 0
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

Bueno, en este caso dentro del statement try, ocurre una excepcion, pero al no estar cubierta por ningun except en el codgio, la excepcion se esparce por el codigo entero y el after run no se ejecutar√°. Lo mas importante a entender es que el finally se ejecuta igualmente haya una excepcion sin cubrir o no.
