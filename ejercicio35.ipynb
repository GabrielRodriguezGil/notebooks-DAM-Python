```python
x = 88 		# Global X

def func():
    x = 99 	# Local X: hides global, but we want this here

func()
print(x) #	 Prints 88: unchanged
```

X se define antes de la funcion con lo cual es una variable global la cual no se ve cambiada por la nueva variable x local, las cuales son variables distintas. Por eso se hace print de x = 88

```python
x = 88 			# Global X

def func():
	global x
	x = 99 		# Global X: outside def

func()
print("x =", x) # Prints 99
```

X se define antes de la funcion como en la anterior pero esta x global es llamada dentro de la función, con lo cual a diferencia de en el anterior codigo, la variable dentro y fuera de la duncon es la misma.

# LEGB rule

```python
# Global names may be referenced within a function without being declared.

y, z = 1, 2         # Global variables in module

def all_global():
	global x        # Declare globals assigned
	x = y + z       # No need to declare y, z: LEGB rule
                    # x existe ahora en el ambito global con valor 3

all_global()
print('x = %d, y = %d, z = %d' % (x, y, z))
```

Ya que las variables y,z se declaran antes del inicio de la funcion estas son globales y pueden ser llamdas dentro de la función sin tener porque declararlas. De esta manera podemos ejecutar procesos con ellas sin meter mas lineas.

```python
x = 99                      # Global scope name: not used

def f1():
    x = 88
    def f2():               # Enclosing def local
        print("x local = ", x)
        def f3():
            print(x)
        f3()    # Reference made in nested def
    f2()                    # f2 is a temporary function that lives only during the execution of (and is visible only to code in) the enclosing f1

f1()                        # Prints 88: enclosing def local
print("x global = ", x)     # salida: X = 88; X sigue valiendo 99

# f2()  NameError: name 'f2' is not defined  No se puede invocar a f2() desde el modulo principal
```

Al llamar a la funcion `f1()` la funcion se ejecuta, y esta hace un print de la x local que es accesible dentro de esa funcion y de `f2()` tambien ya que esa `x(local)` seria como una global para `f2()`. Luego al printar la global funciona igual que siempre printara a la variable `x global` definida al principio.

```python
def f1():
	x = 88                 # enclosing scope
	def f2():
		print("x =", x)    # Remembers x in enclosing def scope
	return f2 	           # Return f2 but don't call it => f1() devuelve el objeto funcion con                              nombre (referencia) f2

action = f1()       # Make, return function => action es ahora una función: action = f2
action() 			# Call it now: prints 88  == f2()

# Functions are objects in Python like everything else, and can be passed back as return values from other functions. 
# Most importantly, f2 remembers the enclosing scope’s x in f1 , even though f1 is no longer active.
```

Al definir action = f1(), estas diciendo que action es igual al contenido dentro de f1 con lo caul al final le estas defiendo action=f2. Al llamar a action, f2 seguira recordando lo que se encontraba en el anterior f1() y de esta manera dara el valor de X igualemente. Esto es porque las funciones dentro de otras funciones pueden acceder a variables de la funcion que los contiene. Y las funciones que "existieron" y tienen sus dependientes siguen guardando en memoria sus "datos"

```python
def maker(n):
	def action(x):		# Make and return action
		return x ** n 	# action retains n from enclosing scope
	return action

f = maker(2)            # Pass 2 to argument n -> return x ** 2
f                    # Pass 3 to x, n remembers "2" -> return 3 ** 2
```

En el caso de que llames a la variable f sin aplicarle parametros lo que pasa es que te dice a que apunta esta variable. De esta manera vemos que apunta a la funcion de dentro de maker, de esta manera validamos que f() en funcion, toma el valor de la segunda funcón.
